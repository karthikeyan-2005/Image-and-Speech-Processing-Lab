# a) Convolutio and correlation of speech signals

```python
from scipy import signal
import numpy as np
import librosa
import matplotlib.pyplot as plt

d_file="C:/Users/GFLAB/lab/image_processing_lab/img_sp_lab/dataset/auds/set1.wav"
d1_file="C:/Users/GFLAB/lab/image_processing_lab/img_sp_lab/dataset/auds/set2.wav"
d,sr=librosa.load(d_file,sr=8000)
d1,sr=librosa.load(d1_file,sr=8000)

plt.subplot(2,2,1)
t=np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d,lw=1)

y=signal.convolve(d,d1)
plt.grid(True)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")

plt.subplot(2,2,2)
t=np.linspace(0,len(d1)/sr,len(d1))
plt.plot(t,d1,lw=1)
plt.grid(True)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")

plt.subplot(2,2,3)
t=np.linspace(0,len(y)/sr,len(y))
plt.plot(t,y,lw=1)
plt.grid(True)
plt.title("Covolution of signals")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.grid(True)
plt.show()
```

# b)Correlation of speech signals

### AutoCorrelation

```python
import numpy as np
import librosa
import matplotlib.pyplot as plt

d_file="C:/Users/GFLAB/lab/image_processing_lab/img_sp_lab/dataset/auds/speech.wav"
d,sr=librosa.load(d_file,sr=8000)
plt.subplot(2,1,1)
t=np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d,lw=1)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
r=librosa.autocorrelate(d,max_size=10000)
plt.subplot(2,1,2)
print(r.shape)
plt.title("Autocorrelation Plot")
plt.xlabel("Lags")
plt.ylabel("Coefficients")
plt.acorr(r,maxlags=20)
print("The Autocorrelation plot for the data is:")
plt.grid(True)
plt.tight_layout(pad=3.0)
plt.show()
```

### Crosscorrelation

```python
import librosa
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
```

```python
d_file="C:/Users/GFLAB/lab/image_processing_lab/img_sp_lab/dataset/auds/sample1.wav"
d1_file="C:/Users/GFLAB/lab/image_processing_lab/img_sp_lab/dataset/auds/sample2.wav"
d,sr=librosa.load(d_file,sr=8000)
d1,sr=librosa.load(d1_file,sr=8000)
print(d)
print(d1)

correlation=np.correlate(d,d1,mode='full')
lags=np.arange(-100,100)
print(correlation)

plt.subplot(2,2,1)
plt.title('speec file1')
t=np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d,lw=1)
plt.grid(True)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.subplot(2,2,2)

plt.title('speech file2')
t=np.linspace(0,len(d1)/sr,len(d1))
plt.plot(t,d1,lw=1)
plt.grid(True)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.subplot(2,2,3)

plt.plot(lags,correlation[0:200])
plt.title('correlation plot')
plt.xlabel('lags')
plt.ylabel('Coefficients')
plt.grid(True)
plt.tight_layout()
plt.show()
```
