```python
import numpy as np
import matplotlib.pyplot as plt
import librosa.core as lc
import librosa.display
```

```python
y, sr = librosa.load("C:/Users/GFLAB/Dropbox/PC/Documents/Image and speech processing/img_sp_lab/dataset/auds/speech.wav",8000)
n_fft = 1024
```

```python
plt.subplot(2,1,1)
plt.title('Original Speech Signal')
plt.plot(y,label="original file") 
plt.xlabel("Time") 
plt.grid(True)
```

```python
fig,ax = plt.subplots()
mag = np.abs(lc.stft(y,n_fft=n_fft,hop_length=10,win_length=40,window='hamming'))
mag1=librosa.amplitude_to_db(mag,ref=np.max)
img=librosa.display.specshow(mag1,sr=sr,x_axis='s',y_axis='linear')
ax.set(title='wideband spectogram')
fig.colorbar(img,ax=ax,format="%+2.f dB")
```

```python
fig,ax=plt.subplots()
mag1 = np.abs(lc.stft(y,n_fft=n_fft,hop_length=100,win_length=400,window='hamming'))
D1=librosa.amplitude_to_db(mag,ref=np.max)
img1=librosa.display.specshow(D1,sr=sr,x_axis='s',y_axis='linear')
ax.set(title='narrowband spectogram')
fig.colorbar(img1,ax=ax,format="%+2.f dB")
plt.tight_layout()
```
