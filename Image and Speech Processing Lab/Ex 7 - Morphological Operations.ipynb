```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread("C:/Users/GFLAB/lab/image_processing_lab/img_sp_lab/dataset/imgs/aa.tiff",0)
kernel=np.ones((5,5),np.uint8)
erosion=cv2.erode(img,kernel,iterations=1)
dilation=cv2.dilate(img,kernel,iterations=1)
plt.subplot(131)
plt.imshow(img)
plt.title("Original Image")
plt.subplot(132)
plt.imshow(erosion)
plt.title("Eroded Image")
plt.subplot(133)
plt.imshow(dilation)
plt.title("Dilated Image")
```
