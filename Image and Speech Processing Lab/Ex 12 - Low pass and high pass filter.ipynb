# Low pass filter and High pass filter

```python
#################### High Pass #####################
import numpy as np
import scipy.signal as sg
import matplotlib.pyplot as plt
import librosa
from scipy.signal import butter,filtfilt
d_file = "C:/Users/GFLAB/Dropbox/PC/Documents/Image and speech processing/img_sp_lab/dataset/auds/set2.wav"
d,sr = librosa.load(d_file,8000)
t = np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d,lw=1)
plt.xlabel("Time(s)")
plt.ylabel('Amplitude')
plt.title('Original speech file')
plt.grid(True)
plt.show()
b,a = sg.butter(4,2000./(sr/2.),'high')
d_fil = sg.filtfilt(b,a,d)
fig,ax = plt.subplots(1,1,figsize=(6,3))
t = np.linspace(0., len(d)/sr,len(d))
ax.plot(t,d,lw=1)
ax.plot(t,d_fil,lw=1)
plt.xlabel('Time(s)')
plt.title('High pass filter - Filtered signal(show in orange colour)')
plt.grid(True)
plt.show()
```

```python
################### Low pas filter ########################
import numpy as np
import scipy.signal as sg
import matplotlib.pyplot as plt
import librosa
from scipy.signal import butter,filtfilt
d_file = "C:/Users/GFLAB/Dropbox/PC/Documents/Image and speech processing/img_sp_lab/dataset/auds/speech.wav"
d,sr = librosa.load(d_file,8000)
print(d)
b,a = sg.butter(4,500./(sr/2.),'low')
d_fil = sg.filtfilt(b,a,d)
fig,ax = plt.subplots(1,1,figsize=(6,3))
t = np.linspace(0.,len(d)/sr,len(d))
ax.plot(t,d,lw=1)
ax.plot(t,d_fil,lw=1)
plt.xlabel('Time(s)')
plt.title('Low pass filter - Filteredsignal(show orange in colour)')
plt.grid(True)
plt.show()
```
