```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
filter = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
image = cv2.imread('C:/Users/GFLAB/Dropbox/PC/Documents/Image and speech processing/img_sp_lab/dataset/imgs/baboon.jpg',1)
sobel_verti_image = cv2.filter2D(image,-1,filter)
sobel_hori_image=cv2.filter2D(image,-1,np.flip(filter.T,axis=0))
plt.imshow(image)
```

```python
plt.subplot(121)
plt.imshow(sobel_verti_image,cmap='gray')
plt.title("Vertical edge")
plt.subplot(122)
plt.imshow(sobel_hori_image,cmap='gray')
plt.title("Horizontal Edge")
plt.show()
```
